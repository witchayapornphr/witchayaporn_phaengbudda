# ข้อมูลที่สนใจ: ช่วงเวลาไลฟ์ขายของที่มีคนดูไลฟ์เยอะที่สุด

##คำอธิบาย
- ข้อมูลคือ: ข้อมูล เกี่ยวกับการไลฟ์ขายของ ของร้านค้าออนไลน์ เช่น เวลาเริ่มไลฟ์ วันของการไลฟ์ ระยะเวลาไลฟ์ จำนวนผู้ชม และยอดขายที่เกิดขึ้นจากการไลฟ์ในแต่ละครั้ง
- เก็บจากไหน:เก็บจากข้อมูลการไลฟ์ขายของ ร้านค้าออนไล์ เช่น TikTok Live,Facebook Live หรือ Shopee Live
- นำไปใช้ทำอะไรได้และประโยชน์ที่ได้:ใช้วิเคราะห์ว่าช่วงเวลาใดของการไลฟ์มีแนสโน้มทำให้มีผู้ชมจำนวนมากและช่วยเพิ่มยอดขาย เพื่อให้ร้านค้าเลือกเวลาไลฟ์ที่เหมาะสมในการวางแผนโปรโมชั่น และเพิ่มโอกาสในการขายสินค้า



```python
data_idea="ช่วงเวลาไลฟ์ขายของที่มีผู้ชมสูงและช่วยเพิ่มยอดขาย"

print("ข้อมูลที่กลุ่มเราเลือก:",data_idea)
print("ใช้ทำนาย/ตัดสินใจ:","ช่วยร้านค้าเลือกช่วงเวลาไลฟ์ที่เหมาะสมเพื่อเพิ่มจำนวนผู้ชมและเพิ่มยอดขาย")
```
